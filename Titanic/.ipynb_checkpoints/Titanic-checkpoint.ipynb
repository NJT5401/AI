```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# loading raw data from csv
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')

y = train_data['Survived']

features = ['Pclass', 'Sex', 'SibSp', 'Parch']
train_data.Sex = train_data.Sex.replace({'male': 1, 'female': 0})
test_data.Sex = test_data.Sex.replace({'male': 1, 'female': 0})
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

y = tf.convert_to_tensor(y)
X = tf.convert_to_tensor(X)
X_test = tf.convert_to_tensor(X_test)
```

```python
lr = 0.01
opt = keras.optimizers.Adam(lr)
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(4,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=opt, loss='binary_crossentropy', metrics='accuracy')
```

```python
# getting learning rates
lre = np.arange(-3, 0, 0.1)
lrs = 10**lre
```

```python
lrs
```

```python
# testing for best lr
lri = {}

for lr in lrs:
    opt = keras.optimizers.Adam(lr)
    model = Sequential()
    model.add(Dense(32, activation='relu', input_shape=(4,)))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=opt, loss='binary_crossentropy', metrics='accuracy')

    hist = model.fit(X, y, epochs=10, batch_size=16)
    
    lri[lr] = hist.history['loss']
```

```python
lrp = []
lossp = []
for lr, losses in lri.items():
    lrp += [lr]
    lossp += [np.mean(losses)]
```

```python
plt.figure(figsize=(15,5))
plt.plot(lrp, lossp)
plt.ylabel('Loss')
plt.xlabel('Learning Rate')
plt.show()
```
